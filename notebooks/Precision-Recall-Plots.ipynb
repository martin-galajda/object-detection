```python
import matplotlib.pyplot as plt

ground_truths_count = 5
predicted_count = 8

def compute_interpolated_precision(p, idx):
    return max(p[idx:])
    

recalls =                 [0   ,  1./5., 1./5., 2./5., 2./5., 2./5., 3./5., 4./5., 4./5.]
precisions =              [1.  ,  1./1., 1./2., 2./3., 2./4., 2./5., 3./6., 4./7., 4./8.]

interpolated_precisions = [compute_interpolated_precision(precisions, idx) for idx in range(len(precisions))]

points = list(zip(recalls, precisions))


fig, (ax1, ax2) = plt.subplots(2)
fig.set_size_inches(8.5, 10.5)

ax1.plot(recalls, precisions, color="red")
ax1.plot(recalls, interpolated_precisions, color="blue", linestyle=':')
ax1.legend(['Precision/recall curve', 'Interpolated Precision/recall curve'])
ax1.set(xlabel='recall', ylabel='precision',
       title='Precision/recall curve')


ax2.plot(recalls, interpolated_precisions, color="red", linestyle='-')
ax2.set(xlabel='recall', ylabel='precision',
       title='AUC for interpolated precision/recall')
ax2.fill_between(recalls, 0, interpolated_precisions, color="slateblue")
ax2.legend(['Interpolated Precision/recall curve', 'AUC'])


fig.savefig("../resources/precision-recall-plot-vertical.png", dpi=100)
plt.show()
```
